```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pprint import PrettyPrinter
```

```python
url = 'https://seed.com.ng/dollar-to-naira-black-market-exchange-rate/'
data = requests.get(url).text
```

```python
print(data)
```

```python
soup = BeautifulSoup(data,'html5lib')
```

```python
pp = PrettyPrinter(indent=2)
pp.pprint(soup.prettify())
```

```python
soup.find_all('tbody')[1].find_all('tr')
```

```python
current_price = pd.DataFrame(columns=['Dollar to naira','Black Market Rate Today'])

for i in soup.find_all('tbody')[0].find_all('tr')[1:]:
    col = i.find_all('td')
    name = col[0].string
    price = col[1].string

    current_price = current_price.append({'Dollar to naira':name,'Black Market Rate Today':price},ignore_index=True)
```

```python
current_price
```

```python
denum_price = pd.DataFrame(columns=['Denomination in Dollar($)','Black Market Exchange Rate Today in Naira'])

for row in soup.find_all('tbody')[1].find_all('tr')[1:]:
    col = row.find_all('td')
    denum = col[0].string
    price = col[1].string

    denum_price = denum_price.append({'Denomination in Dollar($)':denum,'Black Market Exchange Rate Today in Naira':price},ignore_index=True)
```
